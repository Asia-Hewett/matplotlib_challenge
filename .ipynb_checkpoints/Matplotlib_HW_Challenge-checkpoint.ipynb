## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)



# Combine the data into a single dataset
study_mice = pd.merge(mouse_metadata, study_results)

# Display the data table for preview
study_mice
```

```python
# Checking the number of mice.
total_mice = study_mice['Mouse ID'].nunique()
total_mice
```

```python
# Getting the duplicate mouse* by ID number that shows up for Mouse ID and Timepoint. 

problem_mouse = study_mice.loc[study_mice.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
problem_mouse[0]
```

```python
# Optional: Get all the data for the duplicate Mouse ID. 

duplicate_mice_data = study_mice.loc[study_mice.duplicated(subset=['Mouse ID'])]
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mice = study_mice.drop_duplicates(subset=['Mouse ID'], keep='last')
clean_mice = clean_mice[clean_mice['Mouse ID'] != problem_mouse[0]]

# df = df[df.line_race != 0]
clean_mice
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice.count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# SEM is the standard error of the mean
# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean = study_mice.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = study_mice.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variance = study_mice.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
standard_dev = study_mice.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
SEM = study_mice.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

# mean
# median
# variance
# standard_dev
# SEM

# tumor_volume_data = clean_mice.groupby('Drug Regimen')['Tumor Volume (mm3)']
# tumor_volume_data.describe()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

summary_of_mice_stats = pd.DataFrame({
    "Mean Tumor Vol": mean,
    "Median Tumor Vol": median,
    "Variance in Tumor Vol": variance,
    "Standard Deviation": standard_dev,
    "SEM of Sample": SEM
})

summary_of_mice_stats

# Does this question want us to include every mouse in the study or only the data we've cleaned? 
```

```python
# Put treatments into a list for a for loop (and later for plot labels)

drug_treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']


# Create empty list to fill with tumor vol data (for plotting)

tumor_vol_data = []

# 
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in drug_treatments:
    tumor_volume_per = study_mice.loc[study_mice['Drug Regimen'] == drug, 'Tumor Volume (mm3)']  # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_data.append(tumor_volume_per)
    quartiles = tumor_volume_per.quantile([.25, .50, .75])    # add subset 
    lowerq = quartiles[0.25]
    higherq = quartiles[0.75]                                 # Determine outliers using upper and lower bounds
    iqr = higherq - lowerq
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = higherq + (1.5 * iqr)   
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax1 = plt.subplots()
ax1.set_title("Final Tumor Volume")
ax1.set_ylabel("Final Volume Per Drug")
ax1.boxplot(tumor_vol_data)
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
plt.show()

# There are a few outliers to this dataset
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

tumor_vol_df = clean_mice.loc[clean_mice['Drug Regimen'] == 'Capomulin']
avg_tumor_volume = tumor_vol_df.groupby("Mouse ID").mean()


plt.scatter(avg_tumor_volume['Weight (g)'], avg_tumor_volume['Tumor Volume (mm3)'], marker="x", facecolors="orange")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Mouse Weight V.S Capomulin Avg Tumor Vol")


x_values = avg_tumor_volume['Weight (g)']
y_values = avg_tumor_volume['Tumor Volume (mm3)']


(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")


print(f"R Val is {rvalue**2}")
plt.annotate(line_eq,(20,35),fontsize=15,color="black")

```
